## The n-Queens Problem

* __Goal:__ Find an arrangement of $n$ queens on a $n \times n$ chess board so that no queen is on the same row, column or diagonal as any other queen.

* __State space:__ An arrangement of the queens on the board. We restrict the state space to arrangements where there is only a single queen per column. We represent a state as an integer vector $\mathbf{q} = \{q_1, q_2, \dots, q_n\}$, each number representing the row positions of the queens from left to right. We will call a state a "board."

* __Objective function:__ The number of pairwise conflicts (i.e., two queens in the same row/column/diagonal).
The optimization problem is to find the optimal arrangement $\mathbf{q}^*$ of $n$ queens on the board can be written as:

  > minimize: $\mathrm{conflicts}(\mathbf{q})$
  >
  > subject to: $\mathbf{q} \ \text{contains only one queen per column}$

  Note: the constraint (subject to) is enforced by the definition of the state space.

* __Local improvement move:__ Move one queen to a different row in its column.

* __Termination:__ For this problem there is always an arrangement $\mathbf{q}^*$ with $\mathrm{conflicts}(\mathbf{q}^*) = 0$, however, the local improvement moves might end up in a local minimum. 

## Helper functions

````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import colors
import math

np.random.seed(1234)


def random_board(n):
    """Creates a random board of size n x n. Note that only a single queen is placed in each column!"""
    
    return(np.random.randint(0,n, size = n))

def comb2(n): return n*(n-1)//2 # this is n choose 2 equivalent to math.comb(n, 2); // is int division

def conflicts(board):
    """Caclulate the number of conflicts, i.e., the objective function."""

    n = len(board)
    
    horizontal_cnt = [0] * n
    diagonal1_cnt = [0] * 2 * n
    diagonal2_cnt = [0] * 2 * n
    
    for i in range(n):
        horizontal_cnt[board[i]] += 1
        diagonal1_cnt[i + board[i]] += 1
        diagonal2_cnt[i - board[i] + n] += 1
    
    return sum(map(comb2, horizontal_cnt + diagonal1_cnt + diagonal2_cnt))


def show_board(board, cols = ['white', 'gray'], fontsize = 48):  
    """display the board"""
    
    n = len(board)
    
    # create chess board display
    display = np.zeros([n,n])
    for i in range(n):
        for j in range(n):
            if (((i+j) % 2) != 0): 
                display[i,j] = 1
    
    cmap = colors.ListedColormap(cols)
    fig, ax = plt.subplots()
    ax.imshow(display, cmap = cmap, 
              norm = colors.BoundaryNorm(range(len(cols)+1), cmap.N))
    ax.set_xticks([])
    ax.set_yticks([])
    
    # place queens. Note: Unicode u265B is a black queen
    for j in range(n):
        plt.text(j, board[j], u"\u265B", fontsize = fontsize, 
                 horizontalalignment = 'center',
                 verticalalignment = 'center')
    
    print(f"Board with {conflicts(board)} conflicts.")
    plt.show()
````

## Create a board

````python
board = random_board(4)

show_board(board)
print(f"Queens (left to right) are at rows: {board}")
print(f"Number of conflicts: {conflicts(board)}")
````

## Task 1: Steepest-ascend Hill Climbing Search [30 Points]

Calculate the objective function for all local moves (see definition of local moves above) and always choose the best among all local moves. If there are no local moves that improve the objective, then you have reached a local optimum. 

````python
# import whatever packages here
import random
````

### Note about Program Structure
For all of my algorithms, I have a neighbor selection helper function above that returns a neighbor in accordance to the algorithm's specifications. This way I can separate the neighbor selection process from the rest of the algorithm and debug separately as well as keeping the core algorithms cleaner and more readable

````python
def steepest(current):
    current_highest = current
    # select the best neighbor and return
    for i in range(0, len(current)):
        temp = current.copy()
        for j in range(0, len(current)):
            if j != current[i]:
                temp[i] = j
                if conflicts(current_highest) >= conflicts(temp):
                    current_highest = temp.copy()
    return current_highest
    
````

````python
def hill_climbing(problem, verbose = False):
    current = problem
    if verbose:
        show_board(current)
    while True:
        neighbor = steepest(current)
        if conflicts(neighbor) >= conflicts(current):
            if verbose:
                print("final board config")
                show_board(current)
            return current
        if verbose:
            show_board(neighbor)
        current = neighbor
````

````python
board = random_board(4)
result_board = hill_climbing(board, verbose=False)
show_board(result_board)
````

## Task 2: Stochastic Hill Climbing 1 [10 Points]

Chooses randomly from among all uphill moves till you have reached a local optimum.

````python
def stochastic_random(current):
    uphill_moves = []
    for i in range(0, len(current)):
        temp = current.copy()
        for j in range(0, len(current)):
            if j != current[i]:
                temp[i] = j
                if conflicts(temp) < conflicts(current):
                    uphill_moves.append(temp.copy())
    
    if not uphill_moves:    # if uphill neighbors do not exist, simply return the original state
        return current
    return random.choice(uphill_moves)
````

````python
def hill_climbing_stochastic_random(problem, verbose = False):
    current = problem
    if verbose:
        show_board(current)
    while True:
        neighbor = stochastic_random(current)
        if conflicts(neighbor) >= conflicts(current):
            if verbose:
                print("final board config")
                show_board(current)
            return current
        if verbose:
            show_board(neighbor)
        current = neighbor
````

````python
board = random_board(4)
result = hill_climbing_stochastic_random(board, verbose=False)
show_board(result)
````

## Task 3: Stochastic Hill Climbing 2 [20 Points]

A popular version of stochastic hill climbing generates only a single random local neighbor at a time and accept it if it has a better objective function value than the current state. This is very efficient if each state has many possible successor states. This method is called "First-choice hill climbing" in the textbook.

__Notes:__ 

* Detecting local optima is tricky! You can, for example, stop if you were not able to improve the objective function during the last $x$ tries.

````python
def stochastic_first(current):
    index = random.randint(0, len(current) - 1)
    temp = current.copy()
    value = random.randint(0, len(temp) - 1)
    while value == temp[index]:
        value = random.randint(0, len(temp) - 1)
    
    temp[index] = value
    return temp
````

````python
def hill_climbing_stochastic_first(problem, termination_criteria, verbose = False):
    current = problem
    optima_index = 0
    if verbose:
        show_board(current)
    while True:
        neighbor = stochastic_first(current)
        if conflicts(neighbor) < conflicts(current):                                # if neighbor is better than current, accept it
            current = neighbor
            optima_index = 0
            
            if verbose:
                print("new state")
                show_board(neighbor)
        elif optima_index >= termination_criteria:                                  # if neighbor is worse than current but we have reached termination criteria, then terminate program
            if verbose:
                print(f"could not improve for {termination_criteria} times")
                print("final board configuration")
                show_board(current)
            return current
        else:                                                                       # if neighbor is worse than current, increment index value
            optima_index += 1
````

````python
board = random_board(10)
result = hill_climbing_stochastic_first(board, 100, verbose=False)
show_board(result)
````

## Task 4: Hill Climbing Search with Random Restarts [10 Points]

Hill climbing will often end up in local optima. Restart the each of the three hill climbing algorithm up to 100 times with a random board to find a better (hopefully optimal) solution. Note that restart just means to run the algoithm several times starting with a new random board.

````python
def random_restarts(algorithm, board_size, termination_criteria, stochastic_first = 0, verbose = False):
    result = []
    for i in range(0, termination_criteria):
        board = random_board(board_size)
        if stochastic_first != 0:
            result = algorithm(board, stochastic_first)
        else:
            result = algorithm(board)
        if conflicts(result) == 0:
            return result, i
    return result, -100
````

````python
def run_random_restarts(board_size, termination_criteria):
    result_steepest, iteration_steepest = random_restarts(hill_climbing, board_size, termination_criteria)
    print(f"steepest descent took {iteration_steepest + 1} times to find the optimal solution")
    show_board(result_steepest)

    result_random, iteration_random = random_restarts(hill_climbing_stochastic_random, board_size, termination_criteria)
    print(f"stochastic random took {iteration_random + 1} times to find the optimal solution")
    show_board(result_random)

    result_first, iteration_first = random_restarts(hill_climbing_stochastic_first, board_size, termination_criteria, stochastic_first=100)
    print(f"stochastic first choice took {iteration_first + 1} times to find the optimal solution")
    show_board(result_first)
````

````python
run_random_restarts(board_size=8, termination_criteria=100)
````

````python
import time
def return_average_runtime(algorithm, board_size, termination_criteria = 0):
    average_runtime = 0
    for i in range(0,10):
        board = random_board(board_size)
        if termination_criteria != 0:
            start_time = time.time()
            result = algorithm(board, termination_criteria)
            average_runtime += time.time() - start_time
        else:
            start_time = time.time()
            result = algorithm(board)
            average_runtime += time.time() - start_time
    average_runtime /= 10
    
    return average_runtime
````

````python
def return_average_objective_score(algorithm, board_size, termination_criteria = 0):
    average_os = 0
    for i in range(0,10):
        board = random_board(board_size)
        if termination_criteria != 0:
            average_os += conflicts(algorithm(board, termination_criteria))
            
        else:
            average_os += conflicts(algorithm(board))
    average_os /= 10
    
    return average_os
````

## Graduate student advanced task: Simulated Annealing [10 Points]

__Undergraduate students:__ This is a bonus task you can attempt if you like [+5 Bonus Points].

Simulated annealing is a form of stochastic hill climbing that avoid local optima by also allowing downhill moves with a probability proportional to a temperature. The temperature is decreased in every iteration following an annealing schedule. You have to experiment with the annealing schedule (Google to find guidance on this).


1. Implement simulated annealing for the n-Queens problem.
2. Compare the performance with the previous algorithms. 
2. Discuss your choice of annealing schedule.

````python
def simulated_annealing_neighbor_selection(current):
    index = random.randint(0, len(current) - 1)
    temp = current.copy()
    value = random.randint(0, len(temp) - 1)
    while value == temp[index]:
        value = random.randint(0, len(temp) - 1)
    
    temp[index] = value
    return temp
````

````python
def schedule(t):
    return 200 * (0.9 ** t)
````

````python
# reference: https://medium.com/swlh/how-to-implement-simulated-annealing-algorithm-in-python-ab196c2f56a0
# I did not end up using any code from the reference above for final implementation...it looked like their implementation actually was inferior to textbook pseudocode and scheduling equations

def simulated_annealing(problem, verbose = False):
    t = 1

    current = problem
    while True:
        T = schedule(t)
        if T == 0:
            return current
        neighbor = simulated_annealing_neighbor_selection(current)
        if conflicts(neighbor) < conflicts(current):
            current = neighbor
            if verbose:
                print("new state better than current")
                show_board(current)
            
            if conflicts(current) < 1:
                print("final board configuration")
                show_board(current)
                return current
        else:
            # my custom selection probability model that uses current temperature to simulate a "cooling" effect
            selection_probability = np.exp(-(conflicts(neighbor) - conflicts(current))/T)
            probability = np.random.choice(np.arange(0,2), p=[1-selection_probability, selection_probability])
            
            if selection_probability == 1:
                current = neighbor
                if verbose:
                    print("new state worse than current but accepted")
                    show_board(current)
        t += 1
````

````python
board = random_board(60)
result = simulated_annealing(board, verbose=False)
show_board(result)
````

# Notes on Simulated Annealing
My implementation of simulated annealing uses the expoential cooling function with the alpha value of 0.9. The following code snippet gives the probability of selecting a downhill move and executes that probability.
```py
selection_probability = np.exp(-(conflicts(neighbor) - conflicts(current))/T)
probability = np.random.choice(np.arange(0,2), p=[1-selection_probability, selection_probability])
```
After tinkering with the $T_0$ value for some time, I landed on 200. With the probability of selecting a downhill move decreasing with increasing time, my implementation of simulated annealing is guaranteed to find the optimal solution if temperature decrease happens very slowly. However, in the interest of solving the problem in reasonable amounts of time, I opted to allow for suboptimal solutions given that they were good enough. My implementation was able to consistently get less than 2 conflicts for board sizes up to 60 x 60.
